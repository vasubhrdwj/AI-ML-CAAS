Q5)What is Data Structure, why we need one in programs?
A data structure is a way of organizing and storing data in a computer program or system. It defines the way data is arranged, accessed, and manipulated in memory, allowing efficient operations on the data.


### There are several reasons why data structures are important in programs:
<ol>
<li>Efficient data storage: Data structures help optimize the use of memory by providing specific formats for storing different types of data. They ensure that memory is allocated efficiently and that data can be accessed and manipulated quickly.
    </li>
<li>Fast data retrieval: Data structures enable efficient retrieval of data.  Efficient data retrieval is crucial for improving the overall performance of a program.</li>

<li>Effective data organization: Different data structures are designed for specific purposes and operations. For example, linked lists are useful for dynamic data insertion and deletion, while trees are suitable for hierarchical structures and searching algorithms.</li>

<li>Algorithm design: Data structures often go hand in hand with algorithms. Many algorithms are optimized for specific data structures, and understanding the characteristics and capabilities of different data structures helps in designing efficient algorithms. For example, sorting algorithms may vary depending on whether you're working with arrays, linked lists, or trees.</li>
    </ol>
